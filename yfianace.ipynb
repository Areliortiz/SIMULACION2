```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Con este codigo gradicamos precios(s)

pero es mas util trabajar con rendimientos o retornos

P= $\frac{s_{i}-s_{i-1}}{s_{i}}$


```python
tesla=yf.Ticker('TSLA')  # creamos el objeto ticket, Ticker('TSLA'), este se debe escribir como aparece en https://finance.yahoo.com/lookup/
df= tesla.history(start="2023-12-13",interval="1d")# observamos el historial del objeto
df.head() # observamos las primeras filas
tesla_cierre=df['Close']# cerramos el df de tesla

```

```python
tesla_retornos=np.log(tesla_cierre)-np.log(tesla_cierre.shift(1))
tesla_ret= tesla_retornos.dropna()
```

```python
#Graficamos respecrto a la fecha y hora

plt.figure(figsize=(10,5))
plt.plot(tesla_cierre,color="green")
plt.title("Tesla cierre")
plt.xlabel("Fecha")
plt.ylabel("Precio")
plt.grid()
plt.show
```

```python
UBER=yf.Ticker('UBER')
df_UBER= UBER.history(start="2023-12-13",interval="1d")
df_UBER.head()
UBER_cierre=df_UBER['Close']

```

```python
META=yf.Ticker('META')
df_META= META.history(start="2023-12-13",interval="1d")
df_META.head()
META_cierre=df_META['Close']
```

```python
META_retornos=np.log(META_cierre)-np.log(META_cierre.shift(1))
META_ret= META_retornos.dropna()
```

```python
NFLX=yf.Ticker('NFLX')
df_NFLX= NFLX.history(start="2023-12-13",interval="1d")
df_NFLX.head()
NFLX_cierre=df_NFLX['Close']
```

```python
plt.figure(figsize=(10,5))
plt.plot(NFLX_cierre,color="blue")
plt.title("NFLX cierre")
plt.xlabel("Fecha")
plt.ylabel("Precio")
plt.grid()
plt.show
```

```python
#presenta las 3 ultimas graficas en un solo plano

plt.figure(figsize=(10,5))

plt.plot(UBER_cierre, color="blue", label="UBER Close")
plt.plot(META_cierre, color="red", label="META Close")
plt.plot(NFLX_cierre, color="purple", label="NFLX Close")


plt.title("")
plt.xlabel("fecha")
plt.ylabel("precio")
plt.grid()
plt.legend()
plt.show()
```

```python


# Histograma for UBER, META, and NFLX
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.hist(UBER_cierre, bins=30, color='blue', alpha=0.7)
plt.title("Histogram of UBER Closing precios")
plt.xlabel("precio")
plt.ylabel("frecuencia")
plt.grid()

plt.subplot(1, 3, 2)
plt.hist(META_cierre, bins=30, color='red', alpha=0.7)
plt.title("Histogram of META Closing precios")
plt.xlabel("precio")
plt.ylabel("frecuencia")
plt.grid()

plt.subplot(1, 3, 3)
plt.hist(NFLX_cierre, bins=30, color='purple', alpha=0.7)
plt.title("Histogram of NFLX Closing precios")
plt.xlabel("precio")
plt.ylabel("frecuencia")

plt.grid()
plt.tight_layout()
plt.show()
```
