```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
t=0
I=0
lamda=.34
T=60
times=[]
events=[]
while t<=T:
  x_1=rnd.random()
  t=t-(np.log(x_1))/lamda
  if t<=T:
    I+=1
    times.append(t)
    events.append(I)
  x_2=rnd.random()
  if x_2<= (lamda*t)/lamda:
    I=+I+1

plt.step(times,events,where='post')
plt.show()
```
