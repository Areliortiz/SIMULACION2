```python
import sympy as sp
sp.init_printing()
```

```python
S=sp.Function('S')
ode=sp.Eq(sp.Derivative(S(t),t)-c*S(t),0)
ode
```

```python
sol=sp.dsolve(ode,S(t),ics={S(0):10000},constants={c:0.05})
sp.plot(sol.rhs)
sol.subs(constants)
```

```python
sol = sp.dsolve(ode, S(t), ics={S(0): 10000}, constants={c: 0.05})
# Substitute the value of 'c' into the solution before plotting
plot_expr = sol.rhs.subs({c: 0.05})
sp.plot(plot_expr)
sol.subs(constants) # This line is likely unnecessary as 'c' is already substituted
```
